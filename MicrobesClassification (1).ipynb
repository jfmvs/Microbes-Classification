## Pre-processing

```python
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

### Importing the One Hot encoder to transform the categorical data into numeric and slipitting the dataset between features and labels

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()

X = df.drop(["microorganisms"],axis = 1).to_numpy()
y = df.microorganisms
y = y.array.reshape(-1,1)
```

### Implementing PCA to get principal components, reducing dimensionality and also splitting the data

```python
import seaborn as sns
from pylab import rcParams
import matplotlib.pyplot as plt


def plot_correlation(data):
    rcParams['figure.figsize'] = 18, 22
    fig = plt.figure()
    sns.heatmap(data.corr(), annot=True, fmt=".2f")
    plt.show()
    fig.savefig('corr.png')

# plot correlation & densities
plot_correlation(df)
```

```python
# After a lot of test and try, we discovered that 14 is the best number for features

from sklearn.decomposition import PCA
pca = PCA(n_components=14)
X_pca = pca.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, stratify=y)
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size = 0.10, stratify=y_train)
y_train_oh = pd.DataFrame(encoder.fit_transform(y_train).toarray())
y_test_oh = pd.DataFrame(encoder.fit_transform(y_test).toarray())
y_valid_oh = pd.DataFrame(encoder.fit_transform(y_valid).toarray())
```

### Normalizing the training, testing and validation sets

```python
X_train_norm = preprocessing.minmax_scale(X_train)
X_test_norm = preprocessing.minmax_scale(X_test)
X_valid_norm = preprocessing.minmax_scale(X_valid)
```

## Decision Tree Algorithm

#### Testing with Normalization

```python
from sklearn.tree import DecisionTreeClassifier
model_tree_1 = DecisionTreeClassifier(criterion="entropy", max_depth=30)
model_tree_1.fit(X_train_norm, y_train)
y_tree_pca_pred = model_tree_1.predict(X_test_norm)
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_tree_pca_pred))
```

#### Testing with no normalization

```python
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier(criterion="entropy", max_depth=30)
model_tree.fit(X_train, y_train)
y_tree_pred = model_tree.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_tree_pred))
```

#### Aiming better parameters

```python
from sklearn.model_selection import GridSearchCV

depth = list(range(13, 35))
parameters = {
    "criterion": ['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'],
    'max_depth': depth
}

grid_search_tree = GridSearchCV(DecisionTreeClassifier(), 
                    parameters,
                    scoring  = "accuracy",
                    n_jobs= -1, 
                    verbose=3,
                    cv = 5)

grid_search_tree.fit(X_train, y_train)
```

```python
best_params_tree = grid_search_tree.best_params_
print(best_params_tree)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import time

start = time.time()
model_tree = DecisionTreeClassifier(**best_params_tree)
model_tree.fit(X_train, y_train)
y_tree_pca_pred = model_tree.predict(X_test)
stop = time.time()
time_tree = stop - start
print(f"Total time: {time_tree}s")
print("Accuracy validation: ", model_tree.score(X_valid, y_valid))
print("Accuracy test:",model_tree.score(X_test, y_test), '\n')
print(classification_report(y_test, y_tree_pca_pred), '\n')
print("Confusion Matrix Tree : \n", confusion_matrix(y_test, y_tree_pca_pred),"\n")
```
